```python
from tkinter import *
from tkinter import ttk
import pymysql
from tkinter import messagebox

class passenger:
    def __init__(self,root):
        self.root=root
        self.root.title("Railway Management System ")
        self.root.geometry("1350x700+0+0")
        
        
        #let us make frame
        title=Label(self.root,text="Railway Management System",bd=10,relief=GROOVE,font=("times new romam",40,"bold",),bg="yellow",fg="red")
        #fg=foreground color which is the extra colour we give
        #bd =border
        title.pack(side=TOP,fill=X)#fill x helps to fill the title over the entire width
        
        #----------------- VAriable declaration---------
        self.ticket_no_var=StringVar()
        self.name_var=StringVar()
        self.email_var=StringVar()
        self.gender_var=StringVar()
        self.contact_var=StringVar()
        self.dob_var=StringVar()
        
        self.search_by=StringVar()
        self.search_txt=StringVar()
        
        #-----------------Manage Frame------------------
        #-----------------Manage Frame------------------
        Manage_Frame=Frame(self.root,bd=4,relief=RIDGE,bg="crimson")
        Manage_Frame.place(x=20,y=100,width=450,height=580)
        
        m_title=Label(Manage_Frame,text="Manage Railway",bg="crimson",fg="white",font=("times new romam",30,"bold"))
        m_title.grid(row=0,columnspan=2,pady=20)
        
        lbl_ticket=Label(Manage_Frame,text="ticket no.",bg="crimson",fg="white",font=("times new romam",20,"bold"))
        lbl_ticket.grid(row=1,column=0,pady=10,padx=20,sticky='w')
        
        txt_ticket=Entry(Manage_Frame,textvariable=self.ticket_no_var,font=("times new romam",15,"bold"),bd=5,relief=GROOVE)
        txt_ticket.grid(row=1,column=1,pady=10,padx=20,sticky='w')
        
        
        txt_name=Label(Manage_Frame,text="Name",bg="crimson",fg="white",font=("times new romam",20,"bold"))
        txt_name.grid(row=2,column=0,pady=10,padx=20,sticky='w')
        
        txt_Name=Entry(Manage_Frame,textvariable=self.name_var,font=("times new romam",15,"bold"),bd=5,relief=GROOVE)
        txt_Name.grid(row=2,column=1,pady=10,padx=20,sticky='w')
        
        
        lbl_Email=Label(Manage_Frame,text="Email",bg="crimson",fg="white",font=("times new romam",20,"bold"))
        lbl_Email.grid(row=3,column=0,pady=10,padx=20,sticky='w')
        
        txt_Email=Entry(Manage_Frame,textvariable=self.email_var,font=("times new romam",15,"bold"),bd=5,relief=GROOVE)
        txt_Email.grid(row=3,column=1,pady=10,padx=20,sticky='w')
        
        
        lbl_Gender=Label(Manage_Frame,text="Gender",bg="crimson",fg="white",font=("times new romam",20,"bold"))
        lbl_Gender.grid(row=4,column=0,pady=10,padx=20,sticky='w')
        
        combo_gender=ttk.Combobox(Manage_Frame,textvariable=self.gender_var,font=("times new romam",15,"bold"),state='readonly')
        combo_gender['values']=("Male","Female","Other")
        combo_gender.grid(row=4,column=1,pady=10,padx=20,sticky='w')
        
        lbl_DOB=Label(Manage_Frame,text="D.O.B.",bg="crimson",fg="white",font=("times new romam",20,"bold"))
        lbl_DOB.grid(row=5,column=0,pady=10,padx=20,sticky='w')
        
        txt_DOB=Entry(Manage_Frame,textvariable=self.dob_var,font=("times new romam",15,"bold"),bd=5,relief=GROOVE)
        txt_DOB.grid(row=5,column=1,pady=10,padx=20,sticky='w')
        
        lbl_Contact=Label(Manage_Frame,text="Contact",bg="crimson",fg="white",font=("times new romam",20,"bold"))
        lbl_Contact.grid(row=6,column=0,pady=10,padx=20,sticky='w')
        
        txt_Contact=Entry(Manage_Frame,textvariable=self.contact_var,font=("times new romam",15,"bold"),bd=5,relief=GROOVE)
        txt_Contact.grid(row=6,column=1,pady=10,padx=20,sticky='w')
        
        
        lbl_Address=Label(Manage_Frame,text="Address",bg="crimson",fg="white",font=("times new romam",20,"bold"))
        lbl_Address.grid(row=7,column=0,pady=10,padx=20,sticky='w')
        
        self.txt_Address=Text(Manage_Frame,width=30,height=4,font=("",10))
        self.txt_Address.grid(row=7,column=1,pady=10,padx=20,sticky='w')
        
    #------------------------------------Button Frame-----------------------------------------

        btn_Frame=Frame(Manage_Frame,bd=4,relief=RIDGE,bg="crimson")
        btn_Frame.place(x=15,y=500,width=420)
        
        Addbtn=Button(btn_Frame,text="Add",width=10,command=self.add_passengers).grid(row=0,column=0,padx=10,pady=10)
        updatebtn=Button(btn_Frame,text="Update",width=10,command=self.update_data).grid(row=0,column=1,padx=10,pady=10)
        deletebtn=Button(btn_Frame,text="Delete",width=10,command=self.delete_data).grid(row=0,column=2,padx=10,pady=10)
        Clearbtn=Button(btn_Frame,text="Clear",width=10,command=self.clear).grid(row=0,column=3,padx=10,pady=10)
        
        #-----------------Detail Frame-------------------
        Detail_Frame=Frame(self.root,bd=4,relief=RIDGE,bg="crimson")
        Detail_Frame.place(x=500,y=100,width=800,height=580)
        
        
        
        lbl_search=Label(Detail_Frame,text="Search by",bg="crimson",fg="white",font=("times new romam",20,"bold"))
        lbl_search.grid(row=0,column=0,pady=10,padx=20,sticky='w')
        
        combo_search=ttk.Combobox(Detail_Frame,textvariable=self.search_by,width=10,font=("times new romam",15,"bold"),state='readonly')
        combo_search['values']=("ticket_no","Contact")
        combo_search.grid(row=0,column=1,pady=10,padx=20,sticky='w')
        
        
        txt_Search=Entry(Detail_Frame,textvariable=self.search_txt,width=20,font=("times new romam",10,"bold"),bd=5,relief=GROOVE)
        txt_Search.grid(row=0,column=2,pady=10,padx=20,sticky='w')
        
        searchbtn=Button(Detail_Frame,text="Search",width=10,pady=5,command=self.search_data).grid(row=0,column=3,padx=10,pady=10)
        showallbtn=Button(Detail_Frame,text="ShowAll",width=8,pady=5,command=self.fetch_data).grid(row=0,column=4,padx=10,pady=10)
        
        #-------------------- TABLE FRAME ----------------------
        Table_Frame=Frame(Detail_Frame,bd=4,relief=RIDGE,bg="crimson")
        Table_Frame.place(x=10,y=70,width=760,height=500)
        
        scroll_x=Scrollbar(Table_Frame,orient=HORIZONTAL)
        scroll_y=Scrollbar(Table_Frame,orient=VERTICAL)
        self.passenger_table=ttk.Treeview(Table_Frame,columns=("ticket","name","email","gender","contact","dob","address"),xscrollcommand=scroll_x.set,yscrollcommand=scroll_y.set)
        scroll_x.pack(side=BOTTOM,fill=X)
        scroll_y.pack(side=RIGHT,fill=Y)
        scroll_x.config(command=self.passenger_table.xview)
        scroll_y.config(command=self.passenger_table.yview)
        self.passenger_table.heading("ticket",text="ticket_no")
        self.passenger_table.heading("name",text="Name")
        self.passenger_table.heading("email",text="Email")
        self.passenger_table.heading("gender",text="Gender")
        self.passenger_table.heading("contact",text="Contact")
        self.passenger_table.heading("dob",text="D.O.B")
        self.passenger_table.heading("address",text="Address")
        self.passenger_table['show']='headings'
        self.passenger_table.pack()
        self.passenger_table.column("ticket",width=100)
        self.passenger_table.column("name",width=100)
        self.passenger_table.column("email",width=100)
        self.passenger_table.column("gender",width=100)
        self.passenger_table.column("contact",width=100)
        self.passenger_table.column("dob",width=100)
        self.passenger_table.column("address",width=150)
        self.passenger_table.pack(fill=BOTH,expand=1)
        self.passenger_table.bind("<ButtonRelease-1>",self.get_cursor)
        self.fetch_data()
       
    def add_passengers(self):
        if self.ticket_no_var.get()=="" or self.name_var.get()=="":
            messagebox.showerror("error","All fields are required!!")
        else:
            
            con=pymysql.connect(host="localhost", user="root", password="", database="stm")
            cur=con.cursor()
            cur.execute("insert into passengers values (%s, %s, %s,%s, %s, %s,%s)", (self.ticket_no_var.get(),
                                                                                  self.name_var.get(),
                                                                                  self.email_var.get(),
                                                                                  self.gender_var.get(),
                                                                                  self.contact_var.get(),
                                                                                  self.dob_var.get(),
                                                                                  self.txt_Address.get('1.0',END)
                                                                                 ))

            con.commit()
            self.fetch_data()
            self.clear()
            con.close() 
            messagebox.showinfo("Success","Record have been inserted")
    def fetch_data(self):
        con=pymysql.connect(host="localhost", user="root", password="", database="stm")
        cur=con.cursor()
        cur.execute("select * from passengers ")
        rows=cur.fetchall()
        if len(rows)!=0:
            self.passenger_table.delete(*self.passenger_table.get_children())
            for row in rows:
                self.passenger_table.insert('',END,values=row)
            con.commit()
        con.close()
        
    def clear(self):
        self.ticket_no_var.set("")
        self.name_var.set("")
        self.email_var.set("")
        self.gender_var.set("")
        self.contact_var.set("")
        self.dob_var.set("")
        self.txt_Address.delete("1.0",END)
        
    def get_cursor(self,ev):
        cursor_row=self.passenger_table.focus()
        contents=self.passenger_table.item(cursor_row)
        row=contents['values']
        
        self.ticket_no_var.set(row[0])
        self.name_var .set(row[1])
        self.email_var.set(row[2])
        self.gender_var.set(row[3])
        self.contact_var.set(row[4])
        self.dob_var.set(row[5])
        self.txt_Address.delete("1.0", END)
        self.txt_Address.insert(END, row[6])
        
    def update_data(self):
        con=pymysql.connect(host="localhost",user="root", password="", database="stm")
        cur=con.cursor()
        cur.execute("update passengers set name=%s, email=%s, gender=%s, contact=%s, dob=%s, address=%s where ticket_no=%s",(
                                                                               self.name_var.get(),
                                                                               self.email_var.get(),
                                                                               self.gender_var.get(),
                                                                               self.contact_var.get(),
                                                                               self.dob_var.get(),
                                                                               self.txt_Address.get('1.0',END),
                                                                               self.ticket_no_var.get() ))
        con.commit()
        self.fetch_data()
        self.clear()
        con.close()
        
    def delete_data(self):
        con=pymysql.connect(host="localhost", user="root", password="", database="stm")
        cur=con.cursor()
        cur.execute("delete from passengers where ticket_no=%s", self.ticket_no_var.get())
        con.commit()
        con.close()
        self.fetch_data() 
        self.clear()
        
    def search_data(self):
        
        con=pymysql.connect(host="localhost", user="root", password="", database="stm")
        cur=con.cursor()
        cur.execute("select * from passengers where " + str(self.search_by.get())+"="+str(self.search_txt.get()))
        rows=cur.fetchall()
        if len(rows)!=0:
            self.passenger_table.delete(*self.passenger_table.get_children())
            for row in rows:
                self.passenger_table.insert('',END,values=row)
            con.commit()
        con.close()   
        
        
root=Tk()
ob=passenger(root)
root.mainloop()



```
